```python
import numpy as np
from scipy.optimize import curve_fit
from matplotlib import pyplot as plt
```

```python
data = np.loadtxt('integral_grid_resolved.txt')
```

```python
def linear_fit(x, a,b):
    return a+b*x
```

```python
param = curve_fit(linear_fit, data[:, 0], data[:,2], p0 = [0.0, 1e13])
```

```python
param[0,0]
```

```python
plt.plot(data[:, 0], data[:, 1])
plt.plot(data[:, 0], param[0][0]+param[0][1]*data[:,0])
```
